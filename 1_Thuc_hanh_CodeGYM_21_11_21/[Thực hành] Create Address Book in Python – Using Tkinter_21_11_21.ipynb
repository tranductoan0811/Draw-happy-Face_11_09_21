```python
# Viết chương trình cho phép người dùng tạo, xem và xoá địa chỉ, 
# đồng thời cũng có thể reset dữ liệu trong quá trình nhập.
# đầu vào gồm: name, phone, address
```

```python
# Import Module
from tkinter import *
  
# Information List. Khởi tạo biến, nơi lưu danh sách các địa chỉ (tạm gọi là datas ), giá trị ban đầu sẽ là rỗng
datas = []

# Update Information. Khởi tạo hàm cập nhật datas bên ngoài GUI
def update_book():
    select.delete(0,END)
    for n,p,a in datas:
        select.insert(END, n)
"""
Biến select chính là đại diện cho danh sách datas bên ngoài GUI, được khai báo bằng widget select ở bên dưới hàm
Trước khi cập nhật danh sách mới, cần sử dụng hàm delete(start, end) để xoá những items cũ
Dùng for…in để tiến hành tạo vòng lặp lặp qua danh sach địa chỉ datas
Qua mỗi vòng lặp, sử dụng hàm insert(end, row) để thực hiện việc thêm data (là giá trị lặp hiện tại) vào biến select.

"""        
# Add Information. Khởi tạo hàm thêm địa chỉ vào datas
def add():
    global datas
    datas.append([Name.get(),Number.get(),address.get(1.0, "end-1c")])
    update_book()
"""
Sử dụng từ khoá global để hàm hiểu rằng biến datas đang được tương tác là biến toàn cục.
Sử dụng hàm append() để thêm data vào biến datas
Sau khi tương tác với biến datas xong, gọi hàm update_book() để cập nhật lại datas bên ngoài giao diện
"""
  
# View Information. Tạo hàm thực hiện việc xem địa chỉ đã được chọn (tạm gọi là view())
def view():
    Name.set(datas[int(select.curselection()[0])][0])
    Number.set(datas[int(select.curselection()[0])][1])
    address.delete(1.0,"end")
    address.insert(1.0, datas[int(select.curselection()[0])][2])
"""
Đầu tiên chúng ta cần sử dụng hàm curselection() để lấy giá trị trả về của biến select hiện tại 
(kết quả trả về là vị trí index của địa chỉ đã được chọn)
Dùng int() để chuyển vị trí index vừa nhận thành kiểu int 
datas[int(select.curselection()[0])] lúc này sẽ là địa chỉ với vị trí index tương ứng trong datas, 
và vị trí index 0, 1, 2 sẽ tương ứng với trường name, number và address của địa chỉ vừa lấy được
Vì biến Name và Number được khai báo bằng widget Entry() nên sử dụng hàm set() để cập nhật lại giá trị mới
Biến address được khai báo bằng widget Text() nên cần sử dụng hàm delete(1.0, “end”) 
để xoá giá trị cũ trước và dùng hàm insert(1.0, data) để cập nhật lại giá trị mới
"""
  
# Delete Information. Tạo hàm thực hiện việc xoá địa chỉ đã được chọn
def delete():
    del datas[int(select.curselection()[0])]
    update_book()
"""
Đầu tiên chúng ta cần sử dụng hàm curselection() để lấy giá trị trả về của biến select hiện tại
(kết quả trả về là vị trí index của địa chỉ đã được chọn)
Dùng int() để chuyển vị trí index vừa nhận thành kiểu int 
datas[int(select.curselection()[0])] lúc này sẽ là địa chỉ với vị trí index tương ứng trong datas
Sử dụng keyword del để xoá địa chỉ trong datas
Sau khi tương tác với biến datas xong, gọi hàm update_book() để cập nhật lại datas bên ngoài giao diện.
"""
  
def reset():
    Name.set('')
    Number.set('')
    address.delete(1.0,"end")
  
  
# Create Object. Tạo cửa sổ chính ứng dụng GUI
root = Tk()
  
# Set geometry. tuỳ biến thêm giao diện
root.geometry('400x500')
    
# Add Buttons, Label, ListBox. Khai báo biến Name và Number
Name = StringVar()
Number = StringVar()

# Dùng Widget Frame() để khai báo các vùng chứa
frame = Frame()
frame.pack(pady=10) # pady=10 ở đây sẽ quy định khoảng cách giữa các frame (bên phải và bên trái) là 10px
  
frame1 = Frame()
frame1.pack(pady=10)
  
frame2 = Frame()
frame2.pack(pady=10)

"""
Dùng widget Label(), Entry() để khai báo nơi nhập vào của trường văn bản Name và Number, 
đối với Address thì sẽ sử dụng widget Text() để khai báo vì nó cho phép người dùng nhập được văn bản nhiều dòng
"""
Label(frame, text = 'Name', font='arial 12 bold').pack(side=LEFT)
Entry(frame, textvariable = Name,width=50).pack()
  
Label(frame1, text = 'Phone No.', font='arial 12 bold').pack(side=LEFT)
Entry(frame1, textvariable = Number,width=50).pack()
  
Label(frame2, text = 'Address', font='arial 12 bold').pack(side=LEFT)
address = Text(frame2,width=50,height=10)
address.pack()

# Sử dụng widget Button để tạo các button thực hiện việc các gọi hàm

Button(root,text="Add",font="arial 12 bold",command=add).place(x= 100, y=270)
Button(root,text="View",font="arial 12 bold",command=view).place(x= 100, y=310)
Button(root,text="Delete",font="arial 12 bold",command=delete).place(x= 100, y=350)
Button(root,text="Reset",font="arial 12 bold",command=reset).place(x= 100, y=390)

# Tạo giao diện hiển trị danh sách địa chỉ
"""
Sử dụng widget Scrollbar(master, options) để tạo thanh cuộn,  
master ở đây sẽ đại diện cho nơi mà Scrollbar sẽ được áp dụng, 
options sẽ là các key-value đặc tả thêm độ chi tiết của Scrollbar(), 
ortent=VERTICAL sẽ chỉ định thanh cuộn sẽ được đặt thẳng đứng
Sử dụng widget Listbox() để khai báo biến select, 
biến này sẽ mang giá trị thể hiện danh sách datas bên ngoài GUI, 
yscrollcommand=scrollbar.set sẽ đính kèm Listbox() này vào thanh cuộn scroll_bar.
scroll_bar.config(command=select.yview) 
sử dụng để thiết lập phương thức cuộn dọc yview của Listbox() bởi vì Listbox() cần có chế độ xem dọc
"""

scroll_bar = Scrollbar(root, orient=VERTICAL)
select = Listbox(root, yscrollcommand=scroll_bar.set, height=12)
scroll_bar.config(command=select.yview)
scroll_bar.pack(side=RIGHT, fill=Y)
select.place(x=200,y=260)
  
# Execute Tkinter
"""
Gọi vòng lặp sự kiện chính để các hành động có thể diễn ra trên màn hình máy tính của người dùng
"""
root.mainloop()
```
