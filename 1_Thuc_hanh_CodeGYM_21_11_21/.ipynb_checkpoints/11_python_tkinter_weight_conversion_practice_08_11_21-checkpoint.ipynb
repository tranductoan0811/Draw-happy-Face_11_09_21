```python
# Python program to  create a simple GUI 
# weight converter using Tkinter
 
# import thư viện tkinter
from tkinter import *
 

# Create a GUI window
window = Tk()
  
# Function to convert weight
# given in kg to grams, pounds
# and ounces
def from_kg():
     
    # convert kg to gram
    gram = round(float(weight_value.get())*1000,2)
     
    # convert kg to pound
    pound = round(float(weight_value.get())*2.20462,2)
     
    # convert kg to ounce
    ounce = round(float(weight_value.get())*35.274,2)
     
    # Enters the converted weight to
    # the text widget
    t1.delete("1.0", END)
    t1.insert(END,gram)
     
    t2.delete("1.0", END)
    t2.insert(END,pound)
     
    t3.delete("1.0", END)
    t3.insert(END,ounce)
"""
biến gram, pound, ounce
Sử dụng get() để lấy giá trị từ input nhập kg, sử dụng float() để chuyển giá trị về kiểu float, 
nhân giá trị vừa chuyển với 1000 và gán giá trị này.

Sử dụng delete(“1.0”, END) để xoá giá trị hiện tại của trường giá trị gram
Sử dụng insert(END, gram) để cập nhật lại giá trị mới cho trường gram
Sử dụng delete(“1.0”, END) để xoá giá trị hiện tại của trường giá trị pound
Sử dụng insert(END, pound) để cập nhật lại giá trị mới cho trường pound
Sử dụng delete(“1.0”, END) để xoá giá trị hiện tại của trường giá trị ounce
Sử dụng insert(END, ounce) để cập nhật lại giá trị mới cho trường ounce
"""
# Create the Label widgets
weight_lbl  = Label(window, text = "Enter the weight in Kg")
weight_value = StringVar()
weight_ent = Entry(window, textvariable = weight_value)
gram_lbl = Label(window, text = 'Gram')
pound_lbl = Label(window, text = 'Pounds')
ounce_lbl = Label(window, text = 'Ounce')
"""
Sử dụng widget Label() để tạo nhãn thông báo nhập kg và gán giá trị cho biến t1
Sử dụng StringVar() để khai báo biến t2_value, mang giá trị kg cần chuyển đổi
Sử dụng widget Entry() để tạo trường nhập giá trị kg cần chuyển đổi, gán giá trị cho biến t2, 
option textvariable sẽ được gán bằng biến t2_value

Sử dụng widget Label() để tạo nhãn Gram, Pounds và Ounce 
và gán giá trị lần lượt cho các biến e3, e4 và e5
Sử dụng widget Text() để tạo các trường chứa giá trị của gram, pound và ounce 
sau khi chuyển đổi giá trị từ kg, độ cao và độ dài sẽ là 1 và 20
""" 
# Create the Text Widgets
t1 = Text(window, height = 1, width = 20)
t2 = Text(window, height = 1, width = 20)
t3 = Text(window, height = 1, width = 20)
 
# Create the Button Widget
"""
Sử dụng widget Button() để tạo nút gọi hàm chuyển đổi trọng lượng, 
gán giá trị cho biến b1, option command sẽ được gán bằng hàm chuyển đổi trọng lượng
"""
convert_btn = Button(window, text = "Convert", command = from_kg)
 
# grid method is used for placing
# the widgets at respective positions
# in table like structure
weight_lbl.grid(row = 0, column = 0)
weight_ent.grid(row = 0, column = 1)
gram_lbl.grid(row = 1, column = 0)
pound_lbl.grid(row = 1, column = 1)
ounce_lbl.grid(row = 1, column = 2)
t1.grid(row = 2, column = 0)
t2.grid(row = 2, column = 1)
t3.grid(row = 2, column = 2)
convert_btn.grid(row = 0, column = 2)

"""

Sử dụng widget grid() để tiến hành chia vị trí các cột giá trị
"""
# Start the GUI
window.mainloop()
```
